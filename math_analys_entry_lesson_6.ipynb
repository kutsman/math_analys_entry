### 1. Найти производную выражения:

<br>1. $sinx \cdot cosx$
<br>2. $ln(2x+1)^3$
<br>3. $\sqrt{sin^2(ln(x^3))}$
<br>4. $\frac{x^4}{lnx}$

<br>1. $sinx \cdot cosx$

$\frac{d}{dx}[f(x)g(x)] = f(x)\frac{d}{dx}[g(x)] + g(x)\frac{d}{dx}[f(x)]$

$sinx \cdot (-sinx) + cosx \cdot cosx = cos^2x-six^2x = cos2x$

<br>2. $ln(2x+1)^3$

$\frac{1}{(2x+1)^3} \cdot ((2x+1)^3)' = \frac{1}{(2x+1)^3} \cdot 3(2x+1)^2 \cdot 2 = \frac{1}{(2x+1)^3} \cdot 6(2x+1)^2 =\frac{6}{(2x+1)}$

<br>3. $\sqrt{sin^2(ln(x^3))}$

$\frac{1}{2\sqrt{sin^2(ln(x^3))}} \cdot (\sqrt{sin^2(ln(x^3))})' = \frac{1}{2\sqrt{sin^2(ln(x^3))}} \cdot \frac{6cos(ln(x^3)) \cdot sin(ln(x^3))}{x} = \frac{1}{\sqrt{sin^2(ln(x^3))}} \cdot \frac{3cos(ln(x^3)) \cdot sin(ln(x^3))}{x} = \frac{1}{1} \cdot \frac{3cos(ln(x^3))}{x} = \frac{3cos(3ln(x))}{x} $



<br>4. $\frac{x^4}{lnx}$

$\frac{d}{dx}[\frac{f(x)}{g(x)}] = \frac{f(x)'g(x) - g(x)'f(x)}{g^2(x)}$

$\frac{(x^4)'ln(x) - (ln(x))' x^4}{ln^2x} = \frac{4x^3ln(x) - \frac{1}{x} x^4}{ln^2x} = \frac{4x^3ln(x) - x^3}{ln^2x} = x^3\frac{ln(x^4) - 1}{ln^2x} $

### 2. Найти выражение производной функции и ее значение в точке:

<br>$f(x) = cos(x^2+3x), x_0=\sqrt{\pi}$


$f(x)' = -sin(x^2+3x)(x^2+3x)' = -sin(x^2+3x)(2x+3)$

$f(\pi)' = -sin(\pi^2 + 3\pi)(2\pi+3)$

```python
import numpy as np
import math
import matplotlib.pyplot as plt
x = np.arange(-10, 10, 0.1)
plt.plot(x, np.cos(x**2+3*x))
plt.grid(True)
plt.show
```

```python
x = np.arange(-10, 10, 0.1)
plt.plot(x, -np.sin(x**2+3*x)*(2*x+3))
g = np.pi
plt.scatter(np.pi, -np.sin(g**2+3*g)*(2*g+3), c='red', alpha=1)
plt.grid(True)
plt.show()
```

### 3. * Найти значение производной функции в точке:

<br>$f(x)=\frac{x^3-x^2-x-1}{1+2x+3x^2-4x^3}, x_0 = 0$

$\frac{d}{dx}[\frac{f(x)}{g(x)}] = \frac{f(x)'g(x) - g(x)'f(x)}{g^2(x)}$

$f(x)' = \frac{(x^3-x^2-x-1)'(1+2x+3x^2-4x^3)-(x^3-x^2-x-1)(1+2x+3x^2-4x^3)'}{(1+2x+3x^2-4x^3)^2} = \frac{(3x^2-2x-1)(1+2x+3x^2-4x^3)-(x^3-x^2-x-1)(2+6x-12x^2)}{(1+2x+3x^2-4x^3)^2}$

$f(x_0)' = \frac{(-1)(1)-(-1)(2)}{(1)^2} =  \frac{-1+2}{1} = 1$

```python
x = np.arange(-1, 1, 0.001)
y=(-x**4-4*x**2-8*x**2+1)/((1+2*x+3*x**2-4*x**3)**2)
plt.scatter(0, 1, c='orange', alpha=1)
plt.plot(x, y)
plt.grid(True)
plt.show()
```

### 4. Найти угол наклона касательной к графику функции в точке:

<br>$f(x) = \sqrt{3x}\cdot lnx, x_0 = 1$

Геометрический смысл производной - это тангенс угла наклона касательно к графику функции $y = f(x)$ в точке $x_0$

$f(x_0)' = tg\alpha$

Таким образов $f(x)' = (\sqrt{3x})'\cdot lnx + \sqrt{3x}\cdot (lnx)'= \frac{\sqrt{3}}{2\sqrt{x}} \cdot lnx + \sqrt{3x}\cdot \frac{1}{x} = \sqrt{3}$

Отсюда следует: 

$tg\alpha = \sqrt{3}$

ОТВЕТ: угол $\alpha = \frac{\pi}{3}$

$y = kx+b$ уравление касательной

$y = \sqrt{3}x+b = 0$

$b= -\sqrt{3}x, x_0 = 1$, отсюда $b = -\sqrt{3}$

```python
x = np.arange(0, 2, 0.1)
plt.plot(x, np.sqrt(3*x)*np.log(x))
plt.plot(x, np.sqrt(3)*x-np.sqrt(3))
plt.scatter(1, 0, c='green', alpha=1)
plt.grid(True)
plt.show
```
