#### 4. Найти предел:

<br>а) $lim_{x→0}\frac{3x^3-2x^2}{4x^2}$; 
<br>б) $lim_{x→0}\frac{\sqrt{1+x}-1}{\sqrt[3]{1+x}-1}$; 
<br>в) $lim_{x→∞}(\frac{x+3}{x})^{4x+1}$;

б) $lim_{x→0}\frac{\sqrt{1+x}-1}{\sqrt[3]{1+x}-1}$;

$lim_{x→0}\frac{\sqrt{1+x}-1}{\sqrt[3]{1+x}-1} \cdot \frac{\sqrt{1+x}+1}{\sqrt{1+x}+1} \cdot \frac{\sqrt[3]{(1+x)^2}+\sqrt[3]{1+x}+1}{\sqrt[3]{(1+x)^2}+\sqrt[3]{1+x}+1}$;

$lim_{x→0}\frac{\sqrt{(1+x)^2}-\sqrt{1+x} + \sqrt{1+x} - 1}{\sqrt[3]{1+x}-1} \cdot \frac{1}{\sqrt{1+x}+1} \cdot \frac{\sqrt[3]{(1+x)^2}+\sqrt[3]{1+x}+1}{\sqrt[3]{(1+x)^2}+\sqrt[3]{1+x}+1}$;

$lim_{x→0}\frac{\sqrt{(1+x)^2}-\sqrt{1+x} + \sqrt{1+x} - 1}{\sqrt[3]{(1+x)^3}-\sqrt[3]{(1+x)^2}+\sqrt[3]{(1+x)^2}-\sqrt[3]{1+x}+\sqrt[3]{1+x}-1} \cdot \frac{1}{\sqrt{1+x}+1} \cdot \frac{\sqrt[3]{(1+x)^2}+\sqrt[3]{1+x}+1}{1}$;

$lim_{x→0}\frac{1+x-1}{\sqrt[3]{(1+x)^3}-1} \cdot \frac{1}{\sqrt{1+x}+1} \cdot \frac{\sqrt[3]{(1+x)^2}+\sqrt[3]{1+x}+1}{1}$;

$lim_{x→0}\frac{1+x-1}{1+x-1} \cdot \frac{1}{\sqrt{1+x}+1} \cdot \frac{\sqrt[3]{(1+x)^2}+\sqrt[3]{1+x}+1}{1}$;

$lim_{x→0}\frac{1}{\sqrt{1+x}+1} \cdot \frac{\sqrt[3]{(1+x)^2}+\sqrt[3]{1+x}+1}{1}=\frac{1}{\sqrt{1+0}+1} \cdot \frac{\sqrt[3]{(1+0)^2}+\sqrt[3]{1+0}+1}{1}=\frac{1}{1+1} \cdot \frac{1+1+1}{1}=\frac{3}{2}$;

в) $lim_{x→∞}(\frac{x+3}{x})^{4x+1}$; 

 $lim_{x→∞}(1+\frac {3}{x})^{4x+1}=lim_{x→∞}(1+ \frac{1}{\frac {x}{3}})^{4x+1}=lim_{x→∞}(1+ \frac{1}{\frac {x}{3}})^{4x} \cdot (1+ \frac{1}{\frac {x}{3}})=lim_{x→∞}(1+ \frac{1}{\frac {x}{3}})^{ \frac{x}{3} \cdot 12 } \cdot (1+ \frac{1}{\frac {x}{3}}) = e^{12}\cdot  (1 + 0)= e^{12}$

#### 5. Найти предел:

$lim_{x→0}\frac{sin(2x)}{4x}$;
<br>$lim_{x→0}\frac{x}{sin(x)}$;
<br>$lim_{x→0}\frac{x}{arcsin(x)}$;
<br>$lim_{x→∞}(\frac{4x+3}{4x-3})^{6x}$;
<br>$lim_{x→∞}\frac{sinx+lnx}{x}$;
<br>$lim_{x→0}\frac{sinx+lnx}{x}$;

Первым замечательным пределом называется предел отношения синуса бесконечно малой дуги к той же дуге, выраженной в радианной мере:

$lim_{x→0}\frac{sin(x)}{x}=1$;

a) $lim_{x→0}\frac{sin(2x)}{4x}=lim_{x→0}\frac{sin(2x)}{2 \cdot 2x}= 1 \cdot \frac{1}{2}=\frac{1}{2}$;

<br>b) $lim_{x→0}\frac{x}{sin(x)}=lim_{x→0}\frac{1}{\frac{sin(x)}{x}}=1$;

<br>c) $lim_{x→0}\frac{x}{arcsin(x)}=\begin{vmatrix}
t = arcsinx, x = sint\\ 
x→0, t→0
\end{vmatrix}=lim_{t→0}\frac{sint}{t}=1$

Вторым замечательным пределом называется предел:

<br>$lim_{x→∞}(1 + \frac{1}{x})^x = e$;


<br>d) $lim_{x→∞}(\frac{4x+3}{4x-3})^{6x}$;

<br>$lim_{x→∞}(\frac{4+\frac{3}{x}}{4-\frac{3}{x}})^{6x}=lim_{x→∞}(\frac{1+\frac{3}{4x}}{1-\frac{3}{4x}})^{6x}=lim_{x→∞}(\frac{1+\frac{1}{\frac{4x}{3}}}{1-\frac{1}{\frac{4x}{3}}})^{6x}=lim_{x→∞}\frac{(1+\frac{1}{\frac{4x}{3}})^{6x}}{(1-\frac{1}{\frac{4x}{3}})^{6x}}=lim_{x→∞}\frac{(1+\frac{1}{\frac{4x}{3}})^{ \frac{4x}{3} \cdot \frac{9}{2}   }}{(1+\frac{1}{-\frac{4x}{3}})^{  -\frac{4x}{3} \cdot -\frac{9}{2}       }} =\frac { e^{\frac{9}{2}} }  { e^{-\frac{9}{2}}} = \frac { e^{\frac{9}{2}} }  { \frac {1}{e^{\frac{9}{2}}}} =  e^{\frac{9}{2}}  \cdot  e^{\frac{9}{2}} = e^{\frac{9}{2} + \frac{9}{2}} = e^9 $

e) $lim_{x→∞}\frac{sinx+lnx}{x}$;

$lim_{x→∞}\frac{sinx+lnx}{x}=lim_{x→∞}\frac{sinx}{x} + lim_{x→∞}\frac{lnx}{x}= 1 + 0 = 1$;

Применим правило Лопиталя:

$lim_{x→∞}\frac{lnx}{x}=lim_{x→∞}\frac{(lnx)'}{(x)'}=lim_{x→∞}\frac{\frac{1}{x}}{1} = 0 $

<br> f) $lim_{x→0}\frac{sinx+lnx}{x}=lim_{x→0}\frac{sinx}{x} + lim_{x→0}\frac{lnx}{x}$;
