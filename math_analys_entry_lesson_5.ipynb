<br>1) $lim_{x→∞}[ln(x+3)-lnx]$;

<br><br>10) $lim_{x→0}\frac{\sqrt{1+xsinx}-1}{x^2}$; 

Решение: 
    
<br>1) $lim_{x→∞}[ln(x+3)-lnx] = lim_{x→∞} \frac {ln(x+3)}{x+3}(x+3)-\frac{lnx}{x}x  $;

Применим основные свойства:

$ln\frac{x}{y}=lnx-lny$

$a\cdot lnx= lnx^a$

$lim_{x→∞}(1 + \frac{1}{x})^x = e$;

тогда приведем к виду:

$lim_{x→∞}[ln(x+3)-lnx] = lim_{x→∞} ln\frac {x+3}{x} = lim_{x→∞} ln (1 + \frac {3}{x}) = lim_{x→∞} ln (1 + \frac {1}{ \frac{x}{3}})  $;

$ \frac{x}{3} ln (1 + \frac {1}{ \frac{x}{3}}) = ln (1 + \frac {1}{ \frac{x}{3}})^{\frac{x}{3}}$

тогда перепишед $lim$ в виде:

$lim_{x→∞} ln (1 + \frac {1}{ \frac{x}{3}}) = lim_{x→∞} \frac{3}{x} ln (1 + \frac {1}{ \frac{x}{3}})^{\frac{x}{3}} = \frac{3}{∞} \cdot ln(e) = 0 \cdot 1 = 0$;

<br>2) $lim_{x→0}\frac{ln(1+2x)}{arcsin3x}$;

Возьмем производную:

$ln f(x)' = \frac {f'(x)}{f(x)}$

$(acrsinx)' = \frac{1}{\sqrt{1-x^2}} $


$lim_{x→0}\frac{2}{1+2x} \cdot \frac {\sqrt{1-9x^2}}{3} = \frac{2}{1} \cdot \frac{1}{3} = \frac{2}{3}$;






<br>3) $lim_{x→0}\frac{7^x-1}{3^x-1}$;

$(a^x)' = a^xlna, a>0$

Возьмем производную:

$lim_{x→0}\frac{7^xln(7)}{3^xln(3)} =\frac{1\cdot ln(7)}{1 \cdot ln(3)}= \frac{ln(7)}{ln(3)}$;

<br>4) $lim_{a→0}\frac{(x+a)^3-x^3}{a}$;

$lim_{a→0}\frac{(x+a)^3-x^3}{a} = lim_{a→0}\frac{x^3+3x^2a+3xa^2+a^3-x^3}{a} = lim_{a→0}\frac{3x^2a+3xa^2+a^3}{a} = lim_{a→0}3x^2+3xa+a^2  = 3x^2$;


<br>5) $lim_{x→∞}(\frac{x^3}{5x^2+1}-\frac{x^2}{5x-3})$;

$lim_{x→∞}\frac{5x^4-3x^3 - 5x^4-x^2}{(5x^2+1)(5x-3)} = lim_{x→∞}\frac{-3x^3-x^2}{25x^3-15x^2+5x-3} = lim_{x→∞}\frac{x^3(-3-\frac{1}{x})}{x^3(25-15 \frac{1}{x}+5\frac{1}{x^2}-3\frac{1}{x^3})} = lim_{x→∞}\frac{-3-\frac{1}{x}}{25-15 \frac{1}{x}+5\frac{1}{x^2}-3\frac{1}{x^3}} = -\frac{3}{25}$

<br>6) $lim_{x→0}\frac{1-cos4x}{2x \cdot tg2x}$;

Применим правило Лопиталя:


$\frac{1}{2}lim_{x→0}\frac{1-cos4x}{x \cdot tg2x} = \frac{1}{2}lim_{x→0} \frac {(1-cos4x)'}{(xtg2x)'}=\frac{1}{2}lim_{x→0} \frac {0-(-sin4x)4}{1\cdot tg2x+sec^22x \cdot 2\cdot x }=\frac{1}{2}lim_{x→0} \frac {4sin4x}{tg2x+2xsec^22x }$;

Еще раз возьмем производные:

$\frac{1}{2}lim_{x→0} \frac {(4sin4x)'}{(tg2x+2xsec^22x)' }=\frac{1}{2}lim_{x→0} \frac {4cos4x \cdot 4}{sec^22x \cdot 2 +  2 \cdot sec^22x+2\cdot 4xsec^22x\cdot tg2x }=\frac{1}{2}lim_{x→0} \frac {16cos4x}{4sec^22x+8xsec^22x\cdot tg2x }$;

Подставим x=0

$\frac{1}{2}\frac {16cos0}{4\cdot sec^20+8\cdot 0 sec^20\cdot tg0 }=\frac{1}{2}\frac {16\cdot 1}{4\cdot 1+8\cdot 0 sec^20\cdot tg0 }=\frac{1}{2}\frac {16}{4+0}=\frac{1}{2} \cdot 4 = 2$

<br>7) $lim_{x→∞}xsin(\frac{2}{x}) = lim_{x→∞} \frac{sin(\frac{2}{x})}{\frac{1}{x}}$

Применим правило Лопиталя:

$lim_{x→∞} \frac{-\frac{2cos(\frac{2}{x})}{x^2}}{-\frac{1}{x^2}}=lim_{x→∞} 2cos(\frac{2}{x}) = 2 \cdot 1 = 2$

<br>8) $lim_{x→0}(1+tgx)^{ctgx}$;

по формуле $a^x = e^{xln(a)}$ получим:

$lim_{x→0}e^{ctgx \cdot ln(1+tgx)}=lim_{x→0}e^{ln(1+tgx)^{ctgx}}=e^{ln(1+0)^∞}=e^{ln1}=e^0=1$

<br>9) $lim_{x→0}(cos2x)^{\frac{1}{sin^2x}}$

по формуле $a^x = e^{xln(a)}$ получим:

$lim_{x→0}e^{\frac{1}{sin^2x}ln(cos2x)}=lim_{x→0}e^{\frac{1}{sin^2x}ln(1-2sin^2x)}=lim_{x→0}e^{ln(1+(-2sin^2x))^{\frac{-2}{-2sin^2x}}}=lim_{x→0}e^{ln(1+(-2sin^2x))^{\frac{1}{-2sin^2x}\cdot -2}}=e^{ln(e)^{-2}}=e^{-2ln(e)}=e^{-2\cdot 1}=\frac{1}{e^2}$

<br>10) $lim_{x→0}\frac{\sqrt{1+xsinx}-1}{x^2}$;

### 2. Установить характер разрыва функции в точке $x_0$:

<br>1. $f(x)=\frac{x^2-16}{x+4}, x_0=-4;$
<br>2. $f(x)=\frac{sinx}{x}, x_0=0;$

Решение: <br>1. $f(x)=\frac{x^2-16}{x+4}, x_0=-4;$
    
Если в точке $x_0$ существует предел $lim_{x→x_0}f(x_0)$, a $f(x_0)$ не определена, или $lim_{x→x_0}f(x_0)\not=f(x_0)$, то эта точка называется точкой устранимого разрыва.

$f(x_0) = \frac{16-16}{-4+4} = \frac{0}{0}$ - неопределенность

$lim_{x→x_0}\frac{x^2-16}{x+4} = lim_{x→x_0}\frac{(x+4)(x-4)}{x+4}=lim_{x→x_0}\frac{x-4}{1}=lim_{x→x_0}(x-4)= -4-4=-8$

Таким образом в точке $x_0=-4$ - разрыв устранимый

Функция определена при $x<-4$ и при $x>-4$

В точке [-4,-8] - неопределена

```python
import numpy as np
import math
import matplotlib.pyplot as plt
x = np.arange(-20, 20, 0.001)
plt.scatter(4, 0, c='green', alpha=1)
plt.scatter(0, -4, c='green', alpha=1)
plt.scatter(-4, -8, c='red', alpha=1)
plt.plot(x, (x**2-16)/(x+4))
plt.axhline(y=0, color='black', alpha = 0.3) 
plt.axvline(x=0,color='black', alpha = 0.3)
plt.grid(True)
plt.show
```

<br><br>$lim_{x→x_0}\frac{sinx}{x}=lim_{x→0}\frac{sinx}{x} = 1$ - Первый замечательный предел

Таким образом в точке $x_0=0$ - разрыв устранимый

### 3. Исследовать на непрерывность функцию $f(x)$ в точке $x_0$:
<br>1. $f(x)=arctg\frac{2}{x-1}, x_0=1;$
<br>2. $f(x)=\frac{1}{2^{x-3}-1}, x_0=3;$

Решение:
    
<br>1. $f(x)=arctg\frac{2}{x-1}, x_0=1;$

$f(x_0) = arctg\frac{2}{0} = arctg(∞)$

Если в точке $x_0$ не существует хотя бы один из односторонних пределов $lim_{x→x_0+0}f(x)$ и $lim_{x→x_0-0}f(x)$, то точка $x_0$ называется точкой разрыва 2-ого рода.

Если $lim_{x→x_0+0}f(x) - lim_{x→x_0-0}f(x) \not=0$, то $x_0$ - точка скачка функции $f(x)$

$lim_{x→1+} arctg\frac{2}{x-1} = \frac{\pi}{2}$

$lim_{x→1-} arctg\frac{2}{x-1} = -\frac{\pi}{2}$

Функция стремится к -∞ слева и к +∞ справа.

В точке $x_0=1$ точка скачка функции f(x).

```python
print(np.pi/2)
```

```python
x = np.arange(-100, 0.999, 0.001)
x1 = np.arange(1.001, 100, 0.001)
plt.plot(x, np.arctan(2/(x-1)))
plt.plot(x1, np.arctan(2/(x1-1)))
plt.scatter(1, np.pi/2, c='red', alpha=1)
plt.scatter(1, -np.pi/2, c='blue', alpha=1)
plt.text(5, np.pi/2, r'$lim_{x→1+} arctg\frac{2}{x-1}=\frac{\pi}{2}$',color='red')
plt.text(5, -np.pi/2, r'$lim_{x→1-} arctg\frac{2}{x-1}=-\frac{\pi}{2}$',color='blue')
plt.xlim([-50, 50])
plt.ylim([-5,5])
plt.axhline(y=0, color='black', alpha = 0.3) 
plt.axvline(x=0,color='black', alpha = 0.3)
plt.grid(True)
plt.show()
```

Решение:
    
<br>2. $f(x)=\frac{1}{2^{x-3}-1}, x_0=3;$

$f(x_0) = \frac{1}{1-1}= \frac{1}{0}$

Если в точке $x_0$ не существует хотя бы один из односторонних пределов $lim_{x→x_0+0}f(x)$ и $lim_{x→x_0-0}f(x)$, то точка $x_0$ называется точкой разрыва 2-ого рода.

$lim_{x→3+0}\frac{1}{2^{x-3}-1} = +∞$

$lim_{x→3-0}\frac{1}{2^{x-3}-1} = -∞$

Функция имеет разрыв 2-ого рода в точке $x_0=3$

```python
x = np.arange(-100, 2.999, 0.001)
x1 = np.arange(3.001, 100, 0.001)
plt.plot(x, 1/(2**(x-3)-1)   )
plt.plot(x1, 1/(2**(x1-3)-1)   )
plt.axvline(x=3, linestyle ='dotted', color='green')
plt.axhline(y=0, linestyle ='dotted', color='red')
plt.axhline(y=-1, linestyle ='dotted', color='red')
plt.text(-5, 0.5, r'$y=0$',color='red')
plt.text(5, -2, r'$y=-1$',color='red')
plt.text(3, -5, r'$x=3$', color='green', rotation = 90)
plt.xlim([-10, 15])
plt.ylim([-10,10])
plt.axhline(y=0, color='black', alpha = 0.3) 
plt.axvline(x=0,color='black', alpha = 0.3)
plt.grid(True)
plt.show()
```
