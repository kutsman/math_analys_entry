### 1. Найти область определения функции.

$z=\sqrt{1-x^3}+ln(y^2-1)$

<br>$y^2-1 > 0 $
<br>$y\not = 1$ , $y>1$
<br>$y\not = -1$, $y<-1$

<br>$1-x^3 	\ge 0 $ 
<br>$1	\ge x^3 $
<br>$1	\ge x $

Функция определена на $x(-∞;1)$ и $y (-∞; -1)$ $(1;+∞)$

```python
import numpy as np
import math
import matplotlib.pyplot as plt
x = np.arange(-50, 1, 0.01)
y = np.arange(-100, 100, 0.01)
plt.scatter(0, 1, c='red', alpha=1)
plt.plot(x, np.sqrt(1-x**3))
plt.plot(y, np.log(y**2-1))
plt.axhline(y=0, color='black', alpha = 0.3) 
plt.axvline(x=0,color='black', alpha = 0.3)
plt.axvline(x=1, linestyle ='dotted', color='red')
plt.axhline(y=1, linestyle ='dotted', color='red')
plt.axhline(y=-1, linestyle ='dotted', color='red')
plt.xlim([-10, 10])
plt.ylim([-10,10])
plt.grid(True)
plt.show
```

<br>$(x_0;y_0) = (1;1)$
<br>$Az=f'_{x}(x_0;y_0)Ax + f'_{y}(x_0;y_0)Ay$
<br>$dz=z'_{x}dx+z'_{y}dy$

<br>$z=(2xy+cos\frac{x}{y})^{\frac{1}{2}}$
<br>$z'_x=((2xy+cos\frac{x}{y})^{\frac{1}{2}})'=\frac{1}{2}(2xy+cos\frac{x}{y})^{\frac{1}{2}-1}(2xy+cos\frac{x}{y})'=\frac{1}{2}(2xy+cos\frac{x}{y})^{\frac{1}{2}-1}(2y-sin\frac{x}{y} \cdot (\frac{x}{y})')=\frac{1}{2}(2xy+cos\frac{x}{y})^{\frac{1}{2}-1}(2y-sin\frac{x}{y} \cdot \frac{1}{y})=\frac{1}{2}(2xy+cos\frac{x}{y})^{-\frac{1}{2}}(2y-sin\frac{x}{y} \cdot \frac{1}{y})=\frac{2y-sin\frac{x}{y} \cdot \frac{1}{y}}{2\sqrt{2xy+cos\frac{x}{y}}}$

<br>$z'_y=((2xy+cos\frac{x}{y})^{\frac{1}{2}})'=\frac{1}{2}(2xy+cos\frac{x}{y})^{\frac{1}{2}-1}(2xy+cos\frac{x}{y})'=\frac{1}{2}(2xy+cos\frac{x}{y})^{-\frac{1}{2}}(2x-sin\frac{x}{y} \cdot (\frac{x}{y})')=\frac{1}{2}(2xy+cos\frac{x}{y})^{-\frac{1}{2}}(2x-sin\frac{x}{y} \cdot -\frac{x}{y^2})=\frac{1}{2}(2xy+cos\frac{x}{y})^{-\frac{1}{2}}(2x+sin\frac{x}{y} \cdot \frac{x}{y^2})=\frac{2x+sin\frac{x}{y} \cdot \frac{x}{y^2}}{2 \sqrt{2xy+cos\frac{x}{y}}}$
