#### Решение 1:
<br>$x_{n} = n^{(−1)^{n}}$

Распишем последовательность для первых 6 членов:

<br>$x_{1} = 1^{(−1)^{1}}=1^{−1}=\frac{1}{1^1}=1$
<br>$x_{2} = 2^{(−1)^{2}}=2^{1}=2$
<br>$x_{3} = 3^{(−1)^{3}}=3^{−1}=\frac{1}{3}$
<br>$x_{4} = 4^{(−1)^{4}}=4^{1}=4$
<br>$x_{5} = 5^{(−1)^{5}}=5^{−1}=\frac{1}{5}$
<br>$x_{6} = 6^{(−1)^{6}}=6^{1}=6$

мы видим что $n$ в $(-1)^n$ влияет лишь на знак степени "минус" или "плюс" 
<br>и соответственно будет ли член последовательности $x_n$ вычисляться как $n$(четный) или $\frac{1}{n}(нечетный)$

Таким образом вы видим, что при $n→∞$ и (n - четное), 
<br>последовательность $x_{n}$ будет стремится к $∞$ (бесконечности) и она неограничена сверху.

Но если при $n→∞$ n - нечетное, то последовательно $x_{n}$ будет стремится к 0, и значит ограничена снизу.

Не является бесконечно большой, так как ограничена снизу 0 а сверху не ограниченна.

```python
import matplotlib.pyplot as plt
x = []
y = []
def function(n):
    return n ** ((-1)**n)
for n in range(1, 10 + 1):
    x.append(n)
    y.append(function(n))
    print(n, function(n))
```

```python
plt.title(f'N={n}')
plt.plot(x,y)
plt.show
```

```python
for n in range(1, 1000 + 1):
    x.append(n)
    y.append(function(n))
plt.title(f'N={n}')
plt.plot(x,y)
plt.show
```

Решение 3:

$lim_{n→∞}\frac{10n}{n2+1} = lim_{n→∞}\frac{10}{n+\frac{1}{n}}$; 

```python
import matplotlib.pyplot as plt
x2 = []
y2 = []
def function_1(n):
    return 10*n / (n**2 + 1)
for n in range(0, 1000 + 1):
    x2.append(n)
    y2.append(function_1(n))
    #print(n, function_1(n))
```

```python
plt.title(f'N={n}')
plt.plot(x2,y2)
plt.show
```

При $n→∞$ предел стремится к 0

<br>б) $lim_{n→∞}\frac{n^2−n}{n−\sqrt{n}}=lim_{n→∞} \frac {n(n−1)}{n(1−\frac{\sqrt{n}}{n})}=lim_{n→∞} \frac {n−1}{1−\frac{\sqrt{n}}{n}}$; 

здесь сразу видно что при $n→∞$ стремится к $∞$

```python
def function_2(n):
    return (n**2-n)/(n-math.sqrt(n))
def function_3(n):
    return (5*3**n)/(3**n-2)
x3=[]
y3=[]

for n in range(2, 100 + 1):
    x3.append(n)
    y3.append(function_2(n))
    #print(n, function_2(n))
```

```python
plt.title(f'N={n}')
plt.plot(x3,y3)
plt.show
```

При $n→∞$ предел стремится к $∞$

<br>в) $lim_{n→∞}\frac{5\cdot3^n}{3^n−2}=lim_{n→∞}\frac{5\cdot3^n}{3^n(1−\frac{2}{3^n})}=lim_{n→∞}\frac{5}{1−\frac{2}{3^n}}$; 

таким образом при $n→∞$ предел стремится к 5, т.к. $\frac{2}{3^n} → 0$

при n=0 предел имеем $\frac{5}{1−\frac{2}{1}}=\frac{5}{1-2}=\frac{5}{-1}=-5$

```python
x4=[]
y4=[]
for n in range(0, 10 + 1):
    x4.append(n)
    y4.append(function_3(n))
    #print(n, function_3(n))
plt.title(f'N={n}')
plt.plot(x4,y4)
plt.show
```

таким образом при $n→∞$ предел стремится к $5$

Решение 4:
<br><br>4) Найти предел $lim_{n→∞}(\sqrt{n^2+n}-n) =lim_{n→∞}(\sqrt{n^2+n}-n)\cdot (\sqrt{n^2+n}+n) \cdot \frac {1}{\sqrt{n^2+n}+n} = lim_{n→∞}(\sqrt{n^2+n}^2 - n\sqrt{n^2+n} + n\sqrt{n^2+n}-n^2)\cdot \frac {1}{\sqrt{n^2+n}+n}=
lim_{n→∞}(n^2+n-n^2)\cdot \frac {1}{\sqrt{n^2+n}+n}=lim_{n→∞}n\cdot \frac {1}{n\sqrt{1+\frac{1}{n}}+n}=
lim_{n→∞} \frac {1}{\sqrt{1+\frac{1}{n}}+1}$

таким образом при $n→∞$ предел стремится к $0$, так как $\frac{1}{n}→0$ и значит $\sqrt{1+0}+1→2 $
<br><br> И стало быть $lim_{n→∞} \frac {1}{\sqrt{1+\frac{1}{n}}+1}→\frac{1}{2}$
<br><br>таким образом при $n→∞$ предел стремится к $\frac{1}{2}$

```python
x5=[]
y5=[]
def function_4(n):
    return math.sqrt(n**2+n)-n
for n in range(0, 10 + 1):
    x5.append(n)
    y5.append(function_4(n))
    print(n, function_4(n))
plt.title(f'N={n}')
plt.plot(x5,y5)
plt.show
```

```python
epsilon = 10e-7
delta = 1
n = 1
curr = function_4(1)
step = 1
```

```python
%%time
while delta > epsilon:
  next = function_4(n+1)
  delta = abs(curr - next)
  curr = next
  n += step
```

```python
print(n, delta)
print(f"Текущее значение предела: {curr}")
```

так как $cos(n)$ в любом случае будет принимать значения от [-1;1] упростим $lim_{n→∞}\frac{\sqrt{n}}{n+1}=lim_{n→∞}\frac{1}{\frac{n+1}{\sqrt{n}}}=lim_{n→∞}\frac{1}{\sqrt{n}+\frac{1}{\sqrt{n}}}$
