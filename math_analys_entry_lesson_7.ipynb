### 1. Найти интервалы возрастания и убывания функций:

<br>2) $f(x) =xln(x)$

<br>1) $f(x) = x +e^{-x}$

Если функция дифференцируема на интервале $(a;b)$ и для $\forall x\in (a;b): f'(x)>0 (f'(x)<0)$, то f(x) возрастает(убывает) на этом интервале.

<br>$f'(x) = 1-e^{-x}$
<br>$0 = 1-e^{-x}$
<br>$1 = e^{-x}$
<br>$x=0$

<br>$f(x=0) = 0 + 1 = 1$

$f(x)$ возрастает от $(0,+∞)$ и убывает от $(-∞,0)$

```python
import numpy as np
import math
import matplotlib.pyplot as plt
x = np.arange(-50, 50, 0.001)
plt.scatter(0, 1, c='red', alpha=1)
plt.plot(x, x+np.exp(-x))
plt.axhline(y=0, color='black', alpha = 0.3) 
plt.axvline(x=0,color='black', alpha = 0.3)
plt.xlim([-50, 50])
plt.ylim([-50,50])
plt.grid(True)
plt.show
```

### 3. Найти интервалы выпуклости и точки перегиба функций:

<br>1) $f(x) = e^{-x^2}$

Второе достаточно условие точки перегиба $f''(x_0)=0$, a $f'''(x_0)\not =0$

$f'(x)=-2xe^{-x^2}$
<br>$0=-2xe^{-x^2}$
<br>$0=-2x$
<br>$0=x_0$ - экстремум функции
<br>$f''(x)=-2(e^{-x^2}-2x^2e^{-x^2})=-2e^{-x^2}(1-2x^2)$
<br>$0=-2(e^{-x^2}-2x^2e^{-x^2})$
<br>$0=e^{-x^2}-2x^2e^{-x^2}$
<br>$0=e^{-x^2}(1-2x^2)$
<br>$0=1-2x^2$
<br>$1=2x^2$
<br>$\frac{1}{2}=x^2$
<br>$x_1=\sqrt{\frac{1}{2}}$ - точка перегиба
<br>$x_2=-\sqrt{\frac{1}{2}}$ - точка перегиба

<br>$f(x_1)=e^{-\frac{1}{2}}$
<br>$f(x_2)=e^{-\frac{1}{2}}$

<br>$f''(x)$ при $(-∞;-\sqrt{\frac{1}{2}})$ положительна - т.е. выпукла вниз
<br>$f''(x)$ при $(-\sqrt{\frac{1}{2}};\sqrt{\frac{1}{2}})$ отрицательна - т.е. выпукла вверх
<br>$f''(x)$ при $(\sqrt{\frac{1}{2}};+∞($ положительна - т.е. выпукла вниз

```python
x = np.arange(-3, 3, 0.001)
plt.plot(x, np.exp(-x**2))
plt.axhline(y=0, color='black', alpha = 0.3) 
plt.axvline(x=0,color='black', alpha = 0.3)
plt.scatter(0, 1, c='red', alpha=1)
plt.scatter(-np.sqrt(1/2), np.exp(-1/2), c='red', alpha=1)
plt.scatter(np.sqrt(1/2), np.exp(-1/2), c='red', alpha=1)
plt.grid(True)
plt.show
```

<br>2) $f(x) = cosx$

<br>$f'(x) = -sinx$
<br>$0=-sinx$
<br>$x=\pi n$, где n - любое целое число

<br>$f''(x)=-cosx$
<br>$0=-cosx$
<br>$0=cosx$
<br>$x = \frac{\pi}{2}+\pi n$ - точки перегиба, где $n$-любое целое число

<br>$при (-\frac{\pi}{2}-2\pi;\frac{\pi}{2}-\pi;)$ - выпуклая вниз
<br>$при (-\frac{\pi}{2}-\pi;\frac{\pi}{2}+\pi;)$ - выпуклая вверх
<br>$при (\frac{\pi}{2}+\pi;\frac{\pi}{2}+2\pi)$ - выпуклая вниз
<br>$при (\frac{\pi}{2}+2\pi;\frac{\pi}{2}+3\pi)$ - выпуклая вверх

<br>$при (\frac{\pi}{2}+n\pi;\frac{\pi}{2}+(n+1)\pi)$ - выпуклая вверх, когда n -четное
<br>$при (\frac{\pi}{2}+n\pi;\frac{\pi}{2}+(n+1)\pi)$ - выпуклая вниз, когда n -нечетное

<br>$при (\frac{\pi}{2}-n\pi;\frac{\pi}{2}-(n-1)\pi)$ - выпуклая ввниз, когда n -четное
<br>$при (\frac{\pi}{2}-n\pi;\frac{\pi}{2}-(n-1)\pi)$ - выпуклая вверх, когда n -нечетное

```python
x = np.arange(-30, 30, 0.001)
z = np.arange(-10, 10, 1)
plt.plot(x, np.cos(x))
plt.axhline(y=0, color='black', alpha = 0.3) 
plt.axvline(x=0,color='black', alpha = 0.3)
for i in z:
    g=(np.pi)/2+(np.pi)*i
    plt.scatter(g, 0, c='red', alpha=1)
plt.axhline(y=1, linestyle ='dotted', color='green')
plt.axhline(y=-1, linestyle ='dotted', color='green')
plt.grid(True)
plt.show
```

### 4. Найти асимптоты графиков функций:

<br>1)$y=\frac{3x}{x+2}$

<br>1)$y=\frac{3x}{x+2}$

<br>$x=x_0$ - вертикальная ассимптота графика функции $f(x)$, если хотя бы один из пределов $lim_{x→x_0+0}f(x)=∞, lim_{x→x_0-0}f(x)=∞$
<br>$y=kx+b$ - наклонная ассимптота графика функции $f(x)$ 
<br>при x→+∞, если $lim_{x→+∞}(f(x)-(kx+b))=0$
<br>при x→-∞, если $lim_{x→-∞}(f(x)-(kx+b))=0$

<br>$k = lim_{x→+-∞} \frac{f(x)}{x}$
<br>$b = lim_{x→+-∞} [f(x)-kx]$

<br>$y=b$ - горизонтальная ассимптота графика функции f(x) при $x→+-∞$ тогда и только тогда, когда существуют пределы $lim_{x→+-∞}f(x)=b$


<br>$k = lim_{x→+∞} \frac{3}{x+2} = 0$
<br>$b = lim_{x→+∞} [\frac{3x}{x+2}-0]=lim_{x→+∞} \frac{3x}{x+2}=lim_{x→+∞} 3 - \frac{6}{x+2}=3$

<br>$y=3$ - горизонтальная ассимптота

<br>$k = lim_{x→-∞} \frac{3}{x+2} = 0$
<br>$b = lim_{x→-∞} [\frac{3x}{x+2}-0]=lim_{x→-∞} \frac{3x}{x+2}=lim_{x→+∞} 3 - \frac{6}{x+2}=3$

$x=-2$ - разрыв 2-ого рода
<br>$lim_{x→-2+0}\frac{3x}{x+2}=+∞$
<br>$lim_{x→-2-0}\frac{3x}{x+2}=-∞$

<br>$x = -2$ - вертикальная ассимптота

```python
x = np.arange(-30, 30, 0.001)
plt.plot(x, 3*x/(x+2))
plt.axhline(y=0, color='black', alpha = 0.3) 
plt.axvline(x=0, color='black', alpha = 0.3)
plt.scatter(0, 0, c='red', alpha=1)
plt.axhline(y=3, linestyle ='dotted', color='green')
plt.axvline(x=-2, linestyle ='dotted', color='green')
plt.xlim([-30, 30])
plt.ylim([-30,30])
plt.grid(True)
plt.show
```
